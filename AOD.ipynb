# FUNCTIONS

```python
def aod(x,y,col):
    plt.figure(figsize=(10,10))
    plt.xlabel("DATE")
    plt.ylabel("AOD")

    plt.plot(x,y,color=col)
```

```python
def statistics(x):
    print("the min value is", min(data[x]))
    print("the max value is", max(data[x]))
    print("the mean value is", np.mean(data[x]))
    print("the standard deviation value is", np.std(data[x]))
```

```python
plt.figure(figsize=(10,10))
plt.xlabel("DATE")
plt.ylabel("440-870_Angstrom_Exponent α")
plt.plot(data["datetime"], data["440-870_Angstrom_Exponent"],"limegreen")
```

## explaining the graph above
Angstrom exponent (α) being the slope of the ln(AOD) versus ln(wavelength) , 
basically referring to the particle size: the smaller the particle the larger the exponent..
Values of α in the lower polluted layer in metro cities (where aerosols are fresh and recently produced) is generally larger than 1.0 (up to 2.0). Whereas in  the upper atmosphere the α values are less than one.
In general α < 1 refers to coarse mode aerosols
and α > 1 refers to fine mode aerosols  dominance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import matplotlib.ticker as ticker
import datetime


data = pd.read_csv(r'C:/Users/nadia/OneDrive/Desktop/11_Thessaloniki.csv',  parse_dates=["datetime"])

print(data)
```

# FUNCTIONS

```python
def aod(x,y,col):
    plt.figure(figsize=(10,10))
    plt.xlabel("DATE")
    plt.ylabel("AOD")

    plt.plot(x,y,color=col)
```

```python
def statistics(x):
    print("the min value is", min(data[x]))
    print("the max value is", max(data[x]))
    print("the mean value is", np.mean(data[x]))
    print("the standard deviation value is", np.std(data[x]))
```

```python
plt.figure(figsize=(10,10))
plt.xlabel("DATE")
plt.ylabel("440-870_Angstrom_Exponent ")
plt.plot(data["datetime"], data["440-870_Angstrom_Exponent"],"limegreen")
```

## explaining the graph above
Angstrom exponent () being the slope of the ln(AOD) versus ln(wavelength) , 
basically referring to the particle size: the smaller the particle the larger the exponent..
Values of  in the lower polluted layer in metro cities (where aerosols are fresh and recently produced) is generally larger than 1.0 (up to 2.0). Whereas in  the upper atmosphere the  values are less than one.
In general  < 1 refers to coarse mode aerosols
and  > 1 refers to fine mode aerosols  dominance

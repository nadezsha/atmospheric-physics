```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import datetime


data = pd.read_csv(r'C:/Users/nadia/OneDrive/Desktop/11_Thessaloniki.csv',  parse_dates=["datetime"])

print(data)
```

```python
def aod(x,y,col):
    plt.figure(figsize=(10,10))
    plt.xlabel("DATE")
    plt.ylabel("AOD")

    plt.plot(x,y,color=col)
```

```python
aod(data["datetime"], data["AOD_440nm"],"blue")
```

```python
aod(data["datetime"], data["AOD_500nm"],"orchid")
```

```python
aod(data["datetime"], data["AOD_675nm"],"pink")
```

```python
def statistics(x):
    print("the min value is", min(data[x]))
    print("the max value is", max(data[x]))
    print("the mean value is", np.mean(data[x]))
    print("the standard deviation value is", np.std(data[x]))
```
